```python
import pandas as pd
df= pd.DataFrame({'name': ['Chetan', 'Sayak', 'Rachan', 'Vanshaj', 'Zubair', 'Elen', 'Erica', 'Zoey', 'Kate',], 'gender': ['Male','Male','Male','Male','Male', 'Female', 'Female', 'Female', 'Female'], 'age':[50,20,40,30,20,15,90,46,26]}) # Use the correct method name 'DataFrame' and make sure the names and genders are strings.
df.head(9)
```

```python
df.iloc[0:3]
df.iloc[0:3, 1]
df.iloc[0:5,0]
df.iloc[0:5,0:2]
#apply a condition
print(df[df['age'] > 80])
```

```python
#condition to specify female and age above 30
print(df[(df['gender'] == 'Male') & (df['age'] > 30)])
```
